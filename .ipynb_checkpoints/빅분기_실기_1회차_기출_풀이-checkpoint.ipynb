# 작업형 1

## 주어진 데이터셋의 범죄율(crim)에서 상위 10개의 데이터에 대해서 10개 중 최솟값으로 대체한 후, AGE 변수에서 값이 80 이상인 데이터들 중 범죄율의 평균값 구하기
- 보스턴 데이터 https://github.com/selva86/datasets/blob/master/BostonHousing.csv

```python
# 보스턴 데이터 불러오기
import pandas as pd

data = pd.read_csv("./boston.csv")
# crim으로 내림차순 정렬
data = data.sort_values(['crim'], ascending=False)
# crim 상위 10개를 10번째 값으로 대체
data['crim'].iloc[0:10] = data['crim'].iloc[10]
# AGE 변수값이 80이상인 데이터의 crim 평균 구하기 
data[data['age'] >= 80]['crim'].mean()
```

## 주어진 데이터 첫번째 행 부터 순서대로 80%까지의 데이터를 추출 후 'total_bedrooms' 변수의 결측값(NA)을 'total_bedrooms' 변수의 중앙값으로 대체하고 대체 전의 'total_bedrooms' 변수 표준편차값과 대체 후의 'total_bedrooms' 변수 표준편차 값 산출
- 캘리포니아 데이터 https://www.kaggle.com/harrywang/housing?select=housing.csv

```python
data = pd.read_csv('./california.csv')
# 첫번째 행 부터 순서대로 80% 까지의 데이터 추출
data_80_n = len(data)*0.8
data_80 = data.iloc[0:16512,:].copy()

# 대체 전의 표준편차 값
data_80_std_bf = data_80['total_bedrooms'].std()

# 대체 후 표준편차 값 
med = data_80['total_bedrooms'].median()
data_80['total_bedrooms'] = data_80['total_bedrooms'].fillna(med)
data_80_std_af = data_80['total_bedrooms'].std()


print(data_80_std_bf, data_80_std_af)
```

## 주어진 데이터 셋에서 age 컬럼의 이상치를 IQR rule에 의해 탐지하고 이상치들의 합을 구하기(단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함)
- 보험료 데이터 https://www.kaggle.com/mirichoi0218/insurance/version/1

```python
data = pd.read_csv('./insurance.csv')
data
```

```python
mean = data['age'].mean()
std = data['age'].std()
upper = mean + std * 1.5
under = mean - std * 1.5
data[(data['age'] > upper) | (data['age'] < under)]['age'].sum()
```

# 작업형 2
고객 구매 데이터를 사용해서 고객이 주문한 물품이 제 시간에 도착여부(Reached.on.Time_Y.N) 예측
- 데이터 https://www.kaggle.com/prachi13/customer-analytics?select=Train.csv

-----------------------------------------------------원본데이터에서 Train, Test 데이터 셋 추출하는 과정이므로 그대로 진행 --

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def exam_data_load(df, target, id_name="", null_name=""):
    if id_name == "":
        df = df.reset_index().rename(columns={"index": "id"})
        id_name = 'id'
    else:
        id_name = id_name
    
    if null_name != "":
        df[df == null_name] = np.nan
    
    X_train, X_test = train_test_split(df, test_size=0.2, shuffle=True, random_state=2021)
    y_train = X_train[[id_name, target]]
    X_train = X_train.drop(columns=[id_name, target])
    y_test = X_test[[id_name, target]]
    X_test = X_test.drop(columns=[id_name, target])
    return X_train, X_test, y_train, y_test 
    
df = pd.read_csv("./Train.csv")
x_train, x_test, y_train, y_test = exam_data_load(df, 
                                                  target='Reached.on.Time_Y.N', 
                                                  id_name='ID')

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# x_train 원핫인코딩
from pandas import get_dummies
Warehouse_block = pd.get_dummies(x_train['Warehouse_block'], prefix='Warehouse_block', drop_first=True)
Mode_of_Shipment = pd.get_dummies(x_train['Mode_of_Shipment'], prefix='Mode_of_Shipment', drop_first=True)
Product_importance = pd.get_dummies(x_train['Product_importance'], prefix='Product_importance', drop_first=True)
Gender = pd.get_dummies(x_train['Gender'], prefix='Gender', drop_first=True)
```

```python
x_train.drop(columns=['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender'], inplace=True)
x_train = pd.concat([x_train, Warehouse_block, Mode_of_Shipment, Product_importance, Gender], axis=1)
x_train
```

```python
# x_test 원핫인코딩
from pandas import get_dummies
Warehouse_block = pd.get_dummies(x_test['Warehouse_block'], prefix='Warehouse_block', drop_first=True)
Mode_of_Shipment = pd.get_dummies(x_test['Mode_of_Shipment'], prefix='Mode_of_Shipment', drop_first=True)
Product_importance = pd.get_dummies(x_test['Product_importance'], prefix='Product_importance', drop_first=True)
Gender = pd.get_dummies(x_test['Gender'], prefix='Gender', drop_first=True)

x_test.drop(columns=['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender'], inplace=True)
x_test = pd.concat([x_test, Warehouse_block, Mode_of_Shipment, Product_importance, Gender], axis=1)
x_test
```

```python
# x_train 스케일링, fit -> transform
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
```

```python
sc.fit(x_train)
```

```python
x_train_sc = pd.DataFrame(sc.transform(x_train), columns = x_train.columns)
x_train_sc
```

```python
# x_test 스케일링, transform만
x_test_sc = pd.DataFrame(sc.transform(x_test), columns = x_test.columns)
x_test_sc
```

```python
# 분류 모델 생성
# logistic, knn, svc는 sc한 데이터
# tree계열은 sc필요 없음

# 로지스틱 회귀분석
from sklearn.linear_model import LogisticRegression
 
model = LogisticRegression()
model.fit(x_train_sc, y_train['Reached.on.Time_Y.N'])
print('Logistic socre: ',model.score(x_train_sc, y_train))

# Knn
from sklearn.neighbors import KNeighborsClassifier
 
model = KNeighborsClassifier(n_neighbors=4, metric='euclidean')
model.fit(x_train_sc, y_train['Reached.on.Time_Y.N'])
print('KNN socre: ', model.score(x_train_sc, y_train))

# DT
from sklearn.tree import DecisionTreeClassifier
 
model = DecisionTreeClassifier(random_state=1, max_depth=10)
model.fit(x_train, y_train['Reached.on.Time_Y.N'])
print('DTree score: ', model.score(x_train,y_train))

# RF
from sklearn.ensemble import RandomForestClassifier
 
model = RandomForestClassifier(max_depth=10, n_estimators=100)
model.fit(x_train, y_train['Reached.on.Time_Y.N'])
print('RF score: ', model.score(x_train,y_train))
```

```python
# 결과 저장하기
# RF 성능이 가장 좋아서 파라미터로 sc 안한 x_test를 넘김
predict = model.predict_proba(x_test)
```

```python
output = pd.DataFrame({'ID':x_test.index, 'Reached.on.Time_Y.N':predict[:,0]})
```
