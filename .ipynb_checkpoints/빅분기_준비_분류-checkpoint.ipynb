## Raw data를 수정해서: '초등학생_성적.csv'로 저장하기

```python
dat.rename(columns = {'gender':'성별', 'race/ethnicity': '인종', 'parental level of education':'학년', 'lunch':'아침_식사_여부', 'test preparation course':'수강_여부', 
                    'math score':'수학_점수', 'reading score':'IQ', 'writing score':'국어_점수'},
                        inplace = True)
dat
```

```python
# 평균 점수 파생변수 생성
dat['평균_점수'] = (dat['수학_점수'] + dat['국어_점수']) / 2
dat.head()
```

```python
# 4분위 수 구하기
f_standard = dat['평균_점수'].quantile(q=0.25)
s_standard = dat['평균_점수'].quantile(q=0.50)    # 중앙값 기준으로 PASS, FAIL 나누기
t_standard = dat['평균_점수'].quantile(q=0.75)
```

```python
s_standard
```

```python
# def avg_func(x):
#     if x <= f_standard:
#         return 4
#     elif x<= s_standard:
#         return 3
#     elif x<= t_standard:
#         return 2
#     else:
#         return 1

def avg_func(x):
    if x >= s_standard:
        return "PASS"
    else:
        return "FAIL"

dat['합격_여부'] = dat['평균_점수'].apply(avg_func)
dat.head()
```

```python
dat['합격_여부'].value_counts()
```

```python
dat2 = dat[['성별', '학년', '아침_식사_여부', '수강_여부', 'IQ', '국어_점수', '수학_점수', '평균_점수',
            '합격_여부']]

dat2.head()
```

```python
# IQ 데이터 속성값 Range 변환    (60 ~ 140 사이로 변환)
OldMax = dat2.IQ.max()
OldMin = dat2.IQ.min()

NewMax = 140
NewMin = 60

OldRange = (OldMax - OldMin)  
NewRange = (NewMax - NewMin)

def ValueConverter(x):
    NewValue = (((x - OldMin) * NewRange) / OldRange) + NewMin
    return NewValue

dat2.IQ = dat2.IQ.apply(ValueConverter)
dat2.IQ = dat2.IQ.apply(round)

dat2.head()
```

```python
dat2[['IQ']].describe()
```

60에서 140 사이로 잘 들어옴

남자는 수학 점수, 여자는 국어 점수가 높음

```python
dat2.groupby('학년').mean().sort_values('평균_점수')
```

학년이 높을수록 IQ가 높고, 모든 영역에서 점수가 높아짐

- high school : 1학년
- some high school : 2학년
- some college : 3학년

- bachelor's degree : 5학년
- master's degree : 6학년

기준으로 데이터를 변환해보자

```python
# 데이터 변환
dat2['성별'] = dat2['성별'].replace(['female', 'male'], ['여','남'])
dat2['학년'] = dat2['학년'].replace(['high school', 'some high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"], ['1학년','2학년', '3학년', '4학년', '5학년', '6학년'])
```

관련 과목 수강을 하면 점수가 올라간다

```python
# 데이터 변환
dat2['아침_식사_여부'] = dat2['아침_식사_여부'].replace(['free/reduced', 'standard'], [0, 1])
dat2['수강_여부'] = dat2['수강_여부'].replace(['completed', 'none'], [1, 0])
```

## 표준화 or 정규화
수치형 데이터이면, `X`에 대해 표준화 or 정규화 과정을 수행한다.

`IQ` 변수에 대해 시행

```python
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()

# Train set에 대해 스케일링
scaler_minmax.fit(X_train)
X_scaled_minmax_train = scaler_minmax.transform(X_train)

# Test set에 대해 스케일링
scaler_minmax.fit(X_test)
X_scaled_minmax_test = scaler_minmax.transform(X_test)
```

```python
# 확인
pd.DataFrame(X_scaled_minmax_train).describe()
```

```python
# 확인
pd.DataFrame(X_scaled_minmax_test).describe()
```

# 모델 적용 및 평가
- LogisticRegression
- XGBoost
- RandomForest

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

models = [['LogisticRegression', LogisticRegression()], 
          ['RandomForest', RandomForestClassifier()], 
          ['XGBoost',XGBClassifier()],
          ['KNN', KNeighborsClassifier()],
          ['NB', GaussianNB()]
         ]
```

```python
import warnings
warnings.filterwarnings(action='ignore')   # 경고 메시지 무시
```

```python
from sklearn.model_selection import cross_val_score

for name, model in models:
    scores=cross_val_score(model, X_scaled_minmax_train, y_train, cv=5)
    print("[", name, "]")
    print("5개 Train set 정확도:", scores)
    print("Train set 정확도 평균:", scores.mean())
    print("\n")
    
    scores=cross_val_score(model, X_scaled_minmax_test, y_test, cv=5)
    print("5개 Valid set 정확도:", scores)
    print("Valid set 정확도 평균:", scores.mean())
    print("\n")
```

## Confusion Matrix

```python
final_model = XGBClassifier ()
final_model.fit(X_scaled_minmax_train, y_train)
```

```python
# 훈련데이터의 정확도(accuracy) 결과
pred_train=final_model.predict(X_scaled_minmax_train)
final_model.score(X_scaled_minmax_train, y_train)
```

```python
# 테스트데이터의 정확도(accuracy) 결과
pred_test=final_model.predict(X_scaled_minmax_test)
final_model.score(X_scaled_minmax_test, y_test)
```

```python
# 훈련데이터의 혼동행렬 
from sklearn.metrics import confusion_matrix
confusion_train=confusion_matrix(y_train, pred_train)
print("훈련데이터 오차행렬:\n", confusion_train)
```

```python
# 훈련데이터의 분류 모델 평가 결과
from sklearn.metrics import classification_report
cfreport_train=classification_report(y_train, pred_train)
print("분류예측 레포트:\n", cfreport_train)
```

```python
# 테스트데이터의 정확도(accuracy) 결과
confusion_test=confusion_matrix(y_test, pred_test)
print("테스트데이터 오차행렬:\n", confusion_test)
```

```python
# 테스트데이터의 분류 모델 평가 결과
from sklearn.metrics import classification_report
cfreport_test=classification_report(y_test, pred_test)
print("분류예측 레포트:\n", cfreport_test)
```

## 예측 확률 및 roc_auc 스코어 구하기  (Optional)
https://blog.naver.com/PostView.naver?blogId=da0097&logNo=222390408292&categoryNo=0&parentCategoryNo=0

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier

lr = LogisticRegression().fit(X_train, y_train)
probs_lr = lr.predict_proba(X_test)[:,1]
print('Logistic Regression:', roc_auc_score(y_test, probs_lr))
result_lr = roc_auc_score(y_test, probs_lr)

dt = DecisionTreeClassifier().fit(X_train, y_train)
probs_dt = dt.predict_proba(X_test)[:,1]
print('Decision Tree:', roc_auc_score(y_test, probs_dt))
result_dt = roc_auc_score(y_test, probs_dt)

rf = RandomForestClassifier().fit(X_train, y_train)
probs_rf = rf.predict_proba(X_test)[:,1]
print('Random Forest Classifier:', roc_auc_score(y_test, probs_rf))
result_rf = roc_auc_score(y_test, probs_rf)

ada = AdaBoostClassifier().fit(X_train, y_train)
probs_ada = ada.predict_proba(X_test)[:,1]
print('AdaBoost Classifier:', roc_auc_score(y_test, probs_ada))
result_ada = roc_auc_score(y_test, probs_ada)

xg = XGBClassifier().fit(X_train, y_train)
probs_xg = xg.predict_proba(X_test)[:,1]
print('XGB Classifier:', roc_auc_score(y_test, probs_xg))
result_xg = roc_auc_score(y_test, probs_xg)
```

```python
score = pd.DataFrame({'Model':['Logistic Regression', 'Decision Tree',
                               'Random Forest', 'AdaBoost', 'XGB Classifier'],
                      'Score': [result_lr, result_dt, result_rf, result_ada, result_xg]})
score.sort_values(by = 'Score', ascending = False)
score
```
